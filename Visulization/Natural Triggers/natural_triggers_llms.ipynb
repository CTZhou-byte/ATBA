```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = 'Times New Roman'
# 模型名称和数据
models = ['BERT-Large', 'BERT-Base', 'DistilBERT', 'ALBERT-Base', 'GPT-XL', 'GPT-Medium', 
          'GPT-Neo-350', 'GPT-Large', 'OPT-6.7B', 'OPT-350', 'OPT-1.3B', 'OPT-2.7B']


cacc = [91.66, 90.34, 91.76, 87.50, 92.50, 86.94, 89.99, 89.16, 93.88, 92.50, 93.05, 92.50]
asr = [100.0, 81.25, 97.65, 98.43, 100.0, 89.16, 90.00, 72.50, 100.0, 92.50, 96.66, 89.16]


bar_width = 0.35
index = np.arange(len(models))

fig, ax1 = plt.subplots(figsize=(6, 4))

# 绘制 CACC 数据
bars1 = ax1.bar(index - bar_width / 2, cacc, bar_width, label='CACC', color= "#1f78b4")
ax1.set_xlabel('Models', fontsize=15)
ax1.set_ylabel('CACC (%)', color= "#1f78b4")
ax1.tick_params(axis='y', labelcolor= "#1f78b4")
ax1.set_xticks(index)
ax1.set_ylim(80, 96)

# 绘制 ASR 数据
ax2 = ax1.twinx()
bars2 = ax2.bar(index + bar_width / 2, asr, bar_width, label='ASR', color='#B22222', alpha=0.6)
ax2.set_ylabel('ASR (%)', color='#B22222')
ax2.tick_params(axis='y', labelcolor='#B22222')
ax2.set_ylim(70, 101)

# 更新模型名称，并加粗指定的名称
labels = [
    model if model != 'BERT-Large' and model != 'GPT-XL' and model != 'OPT-6.7B' 
    else r'$\mathbf{'+model+'}$' for model in models
]
ax1.set_xticklabels(labels, rotation=45, ha='right')

# 合并图例
lines = [bars1, bars2]
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc='upper right', bbox_to_anchor=(1, 0.2))

plt.tight_layout()
plt.savefig("./natural.pdf", dpi=300)
plt.show()
```
