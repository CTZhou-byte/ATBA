```python
import numpy as np
import matplotlib.pyplot as plt

fontsize = 12
methods = ['Clean', 'ATBA', 'BadNL', 'Sentence']
cacc_bert_large = np.array([92.89, 92.77, 92.32, 93.46])
cacc_bert = np.array([91.84, 90.34, 90.21, 91.03])
cacc_distilbert = np.array([91.84, 91.46, 92.11, 91.84])
cacc_albert = np.array([89.13, 87.50, 90.34, 86.64])
asr = np.array([32.03, 100, 100, 100])
asr_1 = np.array([70.31, 82.03, 19.53, 15.62])
asr_2 = np.array([79.65, 84.37, 18.75, 28.15])
asr_2 = np.array([79.68, 85.15, 12.50, 21.09])

fig, axs = plt.subplots(1, 2, figsize=(8, 3))

bar_width = 0.15  # Width of the bars
bar_positions = np.arange(len(methods))  # Positions for the bars

# CACC comparison
axs[0].bar(bar_positions, cacc_bert_large, width=bar_width, label='BERT-Large', align='center')
axs[0].bar(bar_positions + bar_width, cacc_bert, width=bar_width, label='BERT-Base', align='center')
axs[0].bar(bar_positions + 2 * bar_width, cacc_distilbert, width=bar_width, label='DistilBERT', align='center')
axs[0].bar(bar_positions + 3 * bar_width, cacc_albert, width=bar_width, label='ALBERT', align='center')
axs[0].set_ylabel('CACC (%)', fontsize=fontsize)
axs[0].set_xlabel("(a) Comparison for Clean Accuracy", fontsize=fontsize)
axs[0].set_xticks(bar_positions + bar_width)
axs[0].set_xticklabels(methods)
axs[0].set_ylim(85, 94)
axs[0].legend()

# ASR comparison
axs[1].bar(bar_positions, asr, width=bar_width, label='BERT-Large', align='center')
axs[1].bar(bar_positions + bar_width, asr_1, width=bar_width, label='BERT-Base', align='center')
axs[1].bar(bar_positions + 2 * bar_width, asr_2, width=bar_width, label='DistilBERT', align='center')
axs[1].bar(bar_positions + 3 * bar_width, asr_2, width=bar_width, label='ALBERT', align='center')
axs[1].set_ylabel('ASR (%)', fontsize=fontsize)
axs[1].set_xlabel("(b) Comparison for Attack Success Rate", fontsize=fontsize)
axs[1].set_xticks(bar_positions + bar_width)
axs[1].set_xticklabels(methods)
axs[1].legend()

for ax in axs:
    ax.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.savefig("atp.pdf",dpi=300)
plt.show()
```
